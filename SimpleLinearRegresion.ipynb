```python
# linear_regression_example.py
import matplotlib.pyplot as plt
from scipy import stats

# Data
x = [85, 45, 98, 58, 59, 24, 24, 23, 56, 41]
y = [12, 23, 45, 88, 78, 93, 87, 39, 31, 28]

# Scatter Plot
plt.scatter(x, y)
plt.xlabel("Number of tea drunk")
plt.ylabel("Intelligence quotient")
plt.title("IQ vs Tea Consumption")
plt.grid(True)
plt.show()

# Linear Regression
slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)
def myfunc(x):
    return slope * x + intercept

# Apply regression model
mymodel = list(map(myfunc, x))

# Print regression outputs
print("Slope:", slope)
print("Intercept:", intercept)
print("Predicted y values:", mymodel)
print("Standard Error:", std_err)
print("R-squared:", r_value ** 2)

```

```python
#Plot regression line with data
plt.scatter(x, y)
plt.xlabel("Number of tea drunk")
plt.ylabel("Intelligence quotient")
plt.plot(x, mymodel, color='orange')
plt.title("Regression Line on Scatter Plot")
plt.grid(True)
plt.show()
```
