```python
#Boosting
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix,accuracy_score
from matplotlib.colors import ListedColormap
from sklearn.ensemble import AdaBoostClassifier
import warnings
warnings.filterwarnings('ignore')
```

```python
dataset=pd.read_csv("/content/Logistic_Iris.csv")
dataset.head()
```

```python
x=dataset.iloc[:,[0,1,2,3]].values
x
```

```python
y=dataset.iloc[:,4].values
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
adaboost=AdaBoostClassifier(n_estimators=50,learning_rate=0.2)
adaboost.fit(x_train,y_train)
```

```python
y_pred=adaboost.predict(x_test)
print("pridicted values:")
y_pred
```

```python
adaboost.score(x_test,y_test)
```

```python
acc_single=accuracy_score(y_test,y_pred)*100
print("accuracy of single naive bayes classifier:",acc_single)
```

```python
cm=confusion_matrix(y_test,y_pred)
print("confusion matrix",cm)
```

```python
fig,ax=plt.subplots(figsize=(6,6))
ax.imshow(cm)
ax.grid(False)
ax.xaxis.set(ticks=(0,1,2),ticklabels=('Predicted setosa','Predicted versicolor','Predicted virginica'))
ax.yaxis.set(ticks=(0,1,2),ticklabels=('Actual setosa','Actual versicolor','Actual virginica'))
ax.set_ylim(2.5,-0.5)
for i in range(3):
  for j in range(3):
    ax.text(j,i,cm[i,j],ha='center',va='center',color='white')
plt.show()
```
